# 2. Baseball Database

The [Baseball Database](http://www.seanlahman.com/baseball-archive/statistics/) has an sqlite version. Download it for these exercises.

**2.1** Which player has had the most homeruns?

**2.2** Is there a relation between how many homeruns a player has made in a year and his salary that year? Pull both colums together in a single query



```python
conn1 = sqlite3.connect('data/baseball.db')
c = conn1.cursor()
```

```python
batting = pd.read_csv('data/baseballdatabank-master/core/Batting.csv')
batting.to_sql('Batting', conn1, if_exists='replace', index = False)

Salaries = pd.read_csv('data/baseballdatabank-master/core/Salaries.csv')
Salaries.to_sql('Salaries', conn1, if_exists='replace', index = False)
```

```python
bb1 = pd.read_sql("""SELECT SUM(HR) as HR, playerID
FROM Batting
GROUP BY playerID
ORDER BY HR DESC""", conn1)

bb1
```

```python
bb2 = pd.read_sql("""SELECT salary, HR, s.yearID, b.playerID
FROM salaries s
JOIN batting b
    ON b.playerID = s.playerID
    AND b.yearID = s.yearID
   
ORDER BY HR DESC
LIMIT 10""", conn1)
bb2
```

```python
bb3 = pd.read_sql("""SELECT salary, HR, s.yearID, b.playerID
FROM salaries s
JOIN batting b
    ON b.playerID = s.playerID
    AND b.yearID = s.yearID
   
ORDER BY salary dESC
LIMIT 10""", conn1)
bb3
```

```python
from scipy.stats import linregress

linregress(bb3['HR'], bb3['salary'])
```

P value (0.696477832635459) is more than 0.05 meaning they are not statistically significant.
